````python
single_answer_cols = ['Giới_tính', '2_Loại_da', '3_Tuần_suất_1ngày', '7_Chiết_xuất', '8_Srm_đang_sd', '9_Srm_có_mùi_hương?', '10_Dung_tích', '11_Số_lượng_mua', '12_Chi_tiền_cho_150ml', '13_Tiêu_chí_chọn_srm', '14_Ai_tư_vấn', '18_Tiêu_chí_mua_offline', '19_Tiêu_chí_mua_online', '22_Dùng_thử?', '23_Kỳ_vọng']
df = survey_df[single_answer_cols]
# TODO: code step 1
def calculate_attributes_and_distribution(column):
    num_diff_vals = column.nunique()
    distribution = column.value_counts().to_dict()
    return pd.Series([num_diff_vals, distribution], index=["num_diff_vals", "distribution"])

single_cate_profile = df.agg(calculate_attributes_and_distribution)
single_cate_profile
````

````python
# TODO: code step 2
fig, axes = plt.subplots(3, 5, figsize=(22, 12))
for i, data_dict in enumerate(single_cate_profile.loc['distribution']):
    labels = list(data_dict.keys())
    sizes = list(data_dict.values())
    row, col =  divmod(i, 5)
    pie  = axes[row, col].pie(sizes, labels=None, autopct='%1.1f%%', startangle=90)
    axes[row, col].legend(labels, loc='lower left', bbox_to_anchor=(0, 0), fontsize='small')
    axes[row, col].set_title(single_cate_profile.columns[i])
    
plt.tight_layout()
plt.show()
````

- Next, we will compute the distribution of multiple answer columns. You also perform the following familiar steps:
    - Step #1: For each column, you have to compute the **number of different attributes** and a **dictionary** whose keys and values are attributes and number of each attribute, corresponding. Save your result to a dataframe called `multiple_cate_profile`. Note that in columns `'91_*', '92_*', '93_*'`, there are user-defined values, which are recorded from `Mục khác`. You have to replace these values with `Khác` (you can use `apply` for replacing). For example, column `91_Thích_hương_trái_cây?` has 8 original values (see `./form.pdf`). The rest values will be replace with `Khác`.
    - Step #2: Visualize the 5 values with the highest frequency of occurrence using bar chart. You can use `for` loop to draw the charts.

````python
multi_answer_cols = ['1_Mục_đích_sd_srm','4_Sd_khi_nào','5_Mua_ở_đâu','6_Quảng_cáo_ở_đâu','91_Thích_hương_trái_cây?','92_Thích_hương_rau_củ?','93_Thích_hương_hoa_cỏ?','15_Xuất_xứ','16_Mua_online_ở_đâu','17_Mua_offline_ở_đâu','20_Lo_ngại_về_srm']
df = survey_df[multi_answer_cols]
# TODO: code step 1
list_91 = pd.Series(['Dâu','Táo','Mâm xôi','Cherry','Việt quất','Đào','Cam','Chanh'])
list_92 = pd.Series(['Dưa leo','Thảo mộc','Nha đam','Rau má','Cà chua','Cần tây'])
list_93 = pd.Series(['Lavender','Hoa hồng','Anh đào','Trà xanh','Hoa sứ','Hoa cúc'])
none_differ = ['1_Mục_đích_sd_srm','4_Sd_khi_nào','5_Mua_ở_đâu','6_Quảng_cáo_ở_đâu','15_Xuất_xứ','16_Mua_online_ở_đâu','17_Mua_offline_ở_đâu','20_Lo_ngại_về_srm']
differ = ['91_Thích_hương_trái_cây?','92_Thích_hương_rau_củ?','93_Thích_hương_hoa_cỏ?']
list = pd.Series([list_91,list_92,list_93],index = differ)

def calculate_attributes_and_distribution_mul(column, column_name, list):
    distribution = column.value_counts().to_dict()
    column = column.dropna()
    combined_string = ', '.join(column)
    combined_string = combined_string.replace('?', '')
    attributes = combined_string.split(', ')
    attributes = pd.Series(attributes)
    if column_name in differ:
        attributes = attributes.replace('', pd.NA).dropna()
        attributes[~attributes.isin(list[column_name])] = 'Khác'
    num_diff_vals = attributes.nunique()
    
    attribute_counts = attributes.value_counts()
    if column_name in differ:
        missing_indices = set(list[column_name].values) - set(attribute_counts.index)
        attribute_counts = attribute_counts.append(pd.Series(0, index=missing_indices))
        num_diff_vals = len(attribute_counts.index)
    attribute_counts = attribute_counts.to_dict()
    return pd.Series([num_diff_vals, attribute_counts], index=["num_diff_vals", "distribution"])

multiple_cate_profile = df[multi_answer_cols].agg(
    lambda column: calculate_attributes_and_distribution_mul(column, column_name=column.name, list=list)
)

multiple_cate_profile
````
